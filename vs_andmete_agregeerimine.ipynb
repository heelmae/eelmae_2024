```python
data['distance'] = ((data['koord_y'] - (data['left']+data['right'])/2)**2 + (data['koord_x'] - (data['top']+data['bottom'])/2)**2)**0.5
data.head()
```

```python
data_new = data[['id', 'left', 'right', 'top', 'bottom', 'fid', 'koord_x', 'koord_y', 'isendid', 'distance']]
data_new.head()
```

```python
data_sort_distance = data_new[data_new['id'] == 72092].sort_values(by=['distance'], ascending=True)
```

```python
list_used_farms = data_new[data_new['id'] == 72092]['fid'].unique()
data_new.shape
```

```python
summa = 0
count = 0
kasutatud_farmid = []
for isendite_arv in list_used_farms:
    if summa >= 26000:
        break
    else:
        #juurde = data_new.loc[data_new.j_fid == isendite_arv, 'j_isendid'].values[0]
        juurde = data_sort_distance.loc[data_sort_distance.fid == isendite_arv, 'isendid'].values[0] #Mitu isendit lisame juurde summale
        summa += juurde #Lisame summale isendid juurde
        count += 1 #Loeme mitmest farmist isendeid võtame - EI OLE VAJALIK, lihtsalt kontrollimiseks
        kasutatud_farmid.append(isendite_arv)
        #list_used_farms = np.delete(list_used_farms, np.where(list_used_farms == isendite_arv)) #Kustutame farmide listist ära lisatud farmid
        print("Farmi id " + str(isendite_arv) + " juurde " + str(juurde) + " Summa " + str(summa) + " Count " + str(count))

len(list_used_farms) #Kontrolli palju jäi elemente listi alles
```

```python
# delete all rows from data_new where 'j_fid' is in the list_used_farms
data_new = data_new[~data_new['fid'].isin(kasutatud_farmid)]
data_new.shape
```
